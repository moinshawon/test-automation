## Fill the input search box, and collect desired information from that page. In html, take the element in this way-- id > name > class .. coz id is uniquely identified !!

```python
search = driver.find_element_by_name("s") # find HTML tag input and attribute name = s
search.clear()   # clear the input box first
search.send_keys("test")  # fill the search box 'test'
search.send_keys(Keys.RETURN) # hit enter
time.sleep(2) # just to check what is going on the web page

try:
    main = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "main"))
    ) 
    # the driver will wait for 10 sec ultil it finds the  
    # element id = main
    
    
    articles = main.find_elements_by_tag_name("article")
    # take article tags from main ID. it is a list as i used elements
    # there are lots of articles tag. now i iterate through the loop
    
    for a in articles: # there are many tags named by article so, used loop
        header = a.find_element_by_class_name("entry-summary")
        
        # from each article i collect just one element. that's why
        # there has used element !!
        
        print(header.text) 
    
except: # if you use finally then, after the execution the browser will be close
    driver.quit()
    
```

## Click a link in a page and go to the next pages. Come back again then go forward again !!

```python
driver = webdriver.Chrome(PATH) # creating driver in the path

driver.get('https://www.techwithtim.net/') # go to the site

link1 = driver.find_element_by_link_text("Python Programming")

link1.click()

try:
    link2 = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.LINK_TEXT, "Beginner Python Tutorials"))
    ) 
    # and it has to be clickable text/button. usuage 'a' tag has this.
    
    # search for the text in the page. when we go to new page 
    # wait 10 sec until find that element we are looking for

    
    link2.click()
    time.sleep(2) # just to watch slowly
    
    button = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "sow-button-19310003"))
    )
    # find unique element in the 'a' tag.
    # the text in the button won't work. it has to be in 'a' tag
    
    button.click()
    
    driver.back() # go back
    driver.back()
    time.sleep(2)
    driver.back()
    driver.forward()
    driver.forward() # before clicking the Get started button
    
    
except: # if the text did not find then, close the tab
    driver.close()
    
```

## Action chain for double click, upgrade the value

```python
from selenium.webdriver.common.action_chains import ActionChains

driver = webdriver.Chrome(PATH)
driver.get('https://orteil.dashnet.org/cookieclicker/') # go to the site


driver.implicitly_wait(5) # wait 5 sec before doing next task

cookie = driver.find_element_by_id("bigCookie")
cookie_count = driver.find_element_by_id("cookies")
items = [driver.find_element_by_id("productPrice" + str(i)) for i in range(1,-1,-1)]


action = ActionChains(driver) # creating an instance of ActionChains

action.click(cookie)   # just store in the queue

for a in range(50):
    action.perform() # start the action
    count = int(cookie_count.text.split(" ")[0]) 
    # grab just the number of cookies i made
    
    for i in items:
        value = int(i.text) # coz cant work with element
        if value <= count: # check my created cookies more than the upgrade price
            upgrade_action = ActionChains(driver) 
            # taking new element so need to re-define actionChain
            
            upgrade_action.move_to_element(i) # cursor go to the element
            upgrade_action.click() # store for clicking
            upgrade_action.perform() # now perform all the actions that stored 
            
 
```

## Test case with class, obj, method

```python
# check the test

# from element import BasePageElement
# from locators import MainPageLocators

class SearchTextElement(BasePageElement):
    """This class gets the search text from the specified locator"""

    #The locator for search box where search string is entered
    locator = 'q'


class BasePage(object):
    """Base class to initialize the base page that will be called from all
    pages"""

    def __init__(self, driver):
        self.driver = driver


class MainPage(BasePage):
    """Home page action methods come here. I.e. Python.org"""

    #Declares a variable that will contain the retrieved text
    search_text_element = SearchTextElement()

    def is_title_matches(self):
        """Verifies that the hardcoded text "Python" appears in page title"""

        return "Python" in self.driver.title # return true or false

    def click_go_button(self):
        """Triggers the search"""

        element = self.driver.find_element(*MainPageLocators.GO_BUTTON)
        element.click()


class SearchResultsPage(BasePage):
    """Search results page action methods come here"""

    def is_results_found(self):
        # Probably should search for this text in the specific page
        # element, but as for now it works fine
        return "No results found." not in self.driver.page_source
```

```python
# all the attributes by which we grab the element

from selenium.webdriver.common.by import By

class MainPageLocators(object):
    """A class for main page locators. All main page locators should come here"""

    GO_BUTTON = (By.ID, 'submit')

class SearchResultsPageLocators(object):
    """A class for search results locators. All search results locators should
    come here"""

    pass
```

```python
# send or fill up value in input box
# get the value also

from selenium.webdriver.support.ui import WebDriverWait


class BasePageElement(object):
    """Base page class that is initialized on every page object class."""

    def __set__(self, obj, value):
        """Sets the text to the value supplied"""

        driver = obj.driver
        WebDriverWait(driver, 100).until(
            lambda driver: driver.find_element_by_name(self.locator))
        driver.find_element_by_name(self.locator).clear()
        driver.find_element_by_name(self.locator).send_keys(value)

    def __get__(self, obj, owner):
        """Gets the text of the specified object"""

        driver = obj.driver
        WebDriverWait(driver, 100).until(
            lambda driver: driver.find_element_by_name(self.locator))
        element = driver.find_element_by_name(self.locator)
        return element.get_attribute("value")
```
